```python
#drop the less corelated coloums 
df2=df.drop(["Myalgia","Fever","Headache","Arthralgia","Generalized weakness","watery diarrhea", "Conjunctivitis or Pain behind eyes","coma","IgG","Ecchymosis","Convulsions",], axis = 1)
df2
```

```python
X = df2.drop('Dengue',axis=1)
Y = df2['Dengue']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, random_state=101)
```

```python
Y_train.value_counts()
```

```python
Y_test.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
#from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
# dt=DecisionTreeClassifier(criterion= 'gini', min_samples_split= 100,max_depth=20,max_features='sqrt', splitter= 'best',max_leaf_nodes=10,random_state=42)
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
prediction=dt.predict(X_test)
accuracy_dt=accuracy_score(Y_test,prediction)*100
print("accuracy in DT=",accuracy_dt)
```

```python
print("Accuracy on training set: {:.3f}".format(dt.score(X_train, Y_train)))
print("Accuracy on test set: {:.3f}".format(dt.score(X_test, Y_test)))
```

```python
accuracy_score(prediction,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
#dt=RandomForestClassifier(criterion= 'gini', min_samples_split= 100,max_depth=20,max_features='sqrt', splitter= 'best',max_leaf_nodes=10,random_state=42)

rf=RandomForestClassifier(criterion= 'gini', min_samples_split= 1000,max_leaf_nodes=10)
rf.fit(X_train,Y_train)
```

```python
prediction=rf.predict(X_test)
accuracy_rf=accuracy_score(Y_test,prediction)*100
print("accuracy of RF=",accuracy_rf)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
dani=GradientBoostingClassifier( min_samples_split= 1000)
dani.fit(X_train,Y_train)
```

```python
prediction=dani.predict(X_test)
accuracy_dani=accuracy_score(Y_test,prediction)*100
print("accuracy of GBC",accuracy_dani)
```
